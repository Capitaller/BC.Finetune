# Business Central AL File Statistics

This notebook reads `data/bc_al_text/business_central_al_files.jsonl` and summarizes file-level character counts.

```python
import json
from pathlib import Path
from statistics import mean, median

DATASET_PATH = Path("../data/bc_al_text/business_central_al_files.jsonl")
if not DATASET_PATH.exists():
    DATASET_PATH = Path("data/bc_al_text/business_central_al_files.jsonl")

CHAR_THRESHOLD = 8192

rows = []
with DATASET_PATH.open(encoding="utf-8") as file:
    for line in file:
        if line.strip():
            rows.append(json.loads(line))

if not rows:
    raise ValueError(f"No rows found in {DATASET_PATH}")

for row in rows:
    row["characters"] = int(row.get("characters") or len(row.get("text", "")))

character_counts = [row["characters"] for row in rows]
min_row = min(rows, key=lambda row: row["characters"])
max_row = max(rows, key=lambda row: row["characters"])
files_over_threshold = [row for row in rows if row["characters"] > CHAR_THRESHOLD]

print(f"Dataset path: {DATASET_PATH}")
print(f"Files: {len(rows):,}")
print(f"Total characters: {sum(character_counts):,}")
print(f"Average characters per file: {mean(character_counts):,.1f}")
print(f"Median characters per file: {median(character_counts):,.1f}")
print(f"Minimum characters: {min_row['characters']:,} ({min_row['source']})")
print(f"Maximum characters: {max_row['characters']:,} ({max_row['source']})")
print(f"Files > {CHAR_THRESHOLD:,} characters: {len(files_over_threshold):,} ({len(files_over_threshold) / len(rows):.1%})")
```

## Largest Files

```python
largest_files = sorted(rows, key=lambda row: row["characters"], reverse=True)[:20]

for index, row in enumerate(largest_files, start=1):
    print(f"{index:>2}. {row['characters']:>10,} chars  {row['source']}")
```

## Smallest Files

```python
smallest_files = sorted(rows, key=lambda row: row["characters"])[:20]

for index, row in enumerate(smallest_files, start=1):
    print(f"{index:>2}. {row['characters']:>10,} chars  {row['source']}")
```

## Files Over 8,192 Characters

```python
print(f"Files > {CHAR_THRESHOLD:,} characters: {len(files_over_threshold):,}")

for index, row in enumerate(sorted(files_over_threshold, key=lambda row: row["characters"], reverse=True)[:50], start=1):
    print(f"{index:>2}. {row['characters']:>10,} chars  {row['source']}")
```
